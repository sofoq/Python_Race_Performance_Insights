```python
#step 1 USA races, 50km, 50MI,2020
dfALL[dfALL['Event distance/length'].isin(['50km','50mi'])]
```

```python
dfALL[(dfALL['Event distance/length'].isin(['50km','50mi'])) & (dfALL['Year of event']==2020)]
```

```python
dfALL[(dfALL['Event distance/length'].isin(['50km','50mi'])) & (dfALL['Year of event']==2020) & (dfALL['Athlete country']=='USA')]
```

```python
df=dfALL[(dfALL['Event distance/length'].isin(['50km','50mi'])) & (dfALL['Year of event']==2020) & (dfALL['Athlete country']=='USA')].reset_index(drop=True)
```

```python
df['Event name']=df['Event name'].str.split('(').str.get(0)
```

```python
#Get Athletes Age
df['Athlete Age']=df['Year of event']-df['Athlete year of birth']
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
#Get   Athlete Performance Time
df['Athlete performance']=df['Athlete performance'].str.split(' ').str.get(0)
```

```python
df.head()
```

```python
# Drop Columns
df=df.drop(columns=['Athlete club','Athlete country','Athlete year of birth','Athlete age category'])
```

```python
df[df['Athlete Age'].isna()]
```

```python
df=df.dropna()
```

```python
# check for dublicates
df[df.duplicated()]
```

```python
# fix types
df.dtypes
```

```python
df['Athlete Age']=df['Athlete Age'].astype(int)
```

```python
df['Athlete average speed']=df['Athlete average speed'].astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#rename columns
df.columns
```

```python
df=df.rename(columns={
    'Year of event':'year', 
    'race_date':'race_day', 
    'Event name':'race_name', 
    'Event distance/length':'race_length',
    'Event number of finishers':'number_of_finishers', 
    'Athlete performance':'athlete_performance', 
    'Athlete gender':'athlete_gender',
    'Athlete average speed':'average_speed', 
    'Athlete ID':'id', 
    'Athlete Age':'athlete_age'
  
})
```

```python
df.head(1)
```

```python
df=df[['race_day','race_name','race_length','number_of_finishers','id','athlete_gender','athlete_age','athlete_performance','average_speed']]
```

```python
#difference in speed for the 50km and 50 mi male to female
df.groupby(['race_length','athlete_gender'])['average_speed'].mean()
```

```python
#what age groups are the best in the 50mi race 
df.query('race_length=="50mi"').groupby('athlete_age')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).head(15)
```

```python
#what age groups are the worst in the 50mi race

df.query('race_length=="50mi"').groupby('athlete_age')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).head(15)
```

```python
# compare speed by seasons
df['race_month']=df['race_day'].str.split('.').str.get(1)
```

```python
df.head(1)
```

```python
df['race_month']=df['race_day'].str.split('-').str.get(1).fillna(df['race_day']).str.split('.').str.get(1).astype(int)
```

```python
df.head(10)
```

```python
df['race_season']=df['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df.head(5)
```

```python
race_stats=df.groupby('race_season')['average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
import matplotlib.pyplot as plt


# Creating the line chart
plt.figure(figsize=(10, 5))
sns.lineplot(data=race_stats, x=race_stats.index, y='mean', marker='o', label='Mean Speed')

plt.xlabel('Race Season')
plt.ylabel('Mean Average Speed')
plt.title('Mean Average Speed per Race Season')
plt.xticks(rotation=45)  
plt.legend()
plt.grid(True)
plt.show()
```
